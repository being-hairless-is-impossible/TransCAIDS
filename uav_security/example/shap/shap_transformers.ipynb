```python
import transformers
import shap

# load a transformers pipeline model
model = transformers.pipeline('sentiment-analysis',
                              model = 'ProsusAI/finbert',
                              return_all_scores=True)
```

```python
sententce = '''
The Dow Jones Industrial Average and the other major stock indexes dropped Monday, as Wall Street awaited an afternoon speech from Federal Reserve Chairman Jerome Powell. Early movers on the stock market today were Nvidia (NVDA) and Tesla (TSLA).
'''

model(sententce)
```

```python
explainer = shap.Explainer(model)
shap_values = explainer([sententce])

# visualize the first prediction's explanation for the POSITIVE output class
shap.plots.text(shap_values[0, :, "positive"])
```
